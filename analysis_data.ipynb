```python
import pandas as pd
```

# Train dataset

```python
train_df = pd.read_csv('data/train.csv')
```

```python
train_df.head()
```

```python
len(train_df)
```

```python
train_df['label'].value_counts().plot(kind='bar', title='Distribution of the labels')
```

```python
train_df.text.apply(lambda x: len(x.split())).describe()
```

```python
train_df.text.apply(lambda x: len(x.split())).sum()
```

```python
train_df.text.apply(lambda x: len(x.split())).plot(kind='hist', bins= 100, title='Histogram of the number of words in the text', figsize=(10, 5))
```

```python
train_df.text.apply(lambda x: len(x.split())).plot(kind='box', title='Box plot of the number of words in the text', figsize=(10, 5), ylabel='Number of words')
```
